```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('results.csv')
df.head()
```

```python
data = df.loc[(df['Flows'] == 4) & (df['Packets Per Second']==200)]
data = data.sort_values('Nodes')

metrics = ['Throughput', 'End to End Delay', 'Delivery Ratio', 'Drop Ratio']
ylabels = ['Throughput (Kbps)', 'End to End Delay (microseconds)', 'Delivery Ratio (percentage)',
           'Drop Ratio (percentage)']

for i, m in enumerate(metrics):
    plt.plot(data['Nodes'], data[m])
    plt.scatter(data['Nodes'], data[m])
    plt.xlabel("Number of Nodes")
    plt.ylabel(ylabels[i])
    plt.title(m+" vs Number of Nodes")
    plt.savefig(folder+'/'+m+" vs Number of Nodes"+".png", dpi=1200)
    plt.show()
    print("")
```
