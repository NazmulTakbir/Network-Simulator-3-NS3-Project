```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('results.csv')
data = df.sort_values('Max Range')
data.head()
```

```python
metrics = ['Throughput', 'End to End Delay', 'Delivery Ratio', 'Drop Ratio']
ylabels = ['Throughput (Kbps)', 'End to End Delay (microseconds)', 'Delivery Ratio (percentage)',
           'Drop Ratio (percentage)']

for i, m in enumerate(metrics):
    plt.plot(data['Max Range'], data[m])
    plt.scatter(data['Max Range'], data[m])
    plt.xlabel("Max Range")
    plt.ylabel(ylabels[i])
    plt.title(m+" vs Max Range")
    plt.savefig(folder+'/'+m+" vs Max Range"+".png", dpi=1200)
    plt.show()
    print("")
```
